\w

의미: 단어 문자 (영문자, 숫자, 밑줄 포함)

```python
import re
text = "Hello_world123"
match = re.findall(r'\w+', text)
print(match)  # ['Hello_world123']
```

\W

의미: 비단어 문자 (공백, 특수문자 등)

```python
text = "Hello, world!"
match = re.findall(r'\W+', text)
print(match)  # [', ', ' ']
```

\d

의미: 숫자 (0-9)

```python
text = "There are 2 apples and 3 oranges."
match = re.findall(r'\d+', text)
print(match)  # ['2', '3']
```

\D

의미: 비숫자 (숫자가 아닌 문자)

```python
text = "123abc456"
match = re.findall(r'\D+', text)
print(match)  # ['abc']
```

\s

의미: 공백 문자 (스페이스, 탭, 줄 바꿈 등)

```python
text = "Hello   world"
match = re.findall(r'\s+', text)
print(match)  # ['   ']
```

\S: 비공백 문자 (공백이 아닌 모든 문자)

```python
text = "Hello world"
match = re.findall(r'\S+', text)
print(match)  # ['Hello', 'world']
```

.

의미: 임의의 문자 (개행 문자를 제외한 모든 문자)

```python
text = "최소영"
match = re.findall(r'최.영', text)
print(match)  # ['a1b']
```

^

의미: 문자열의 시작


```python
text = "Hello world"
match = re.findall(r'^Hello', text)
print(match)  # ['Hello']
```

$

의미: 문자열의 끝

```python
text = "Hello world."
match = re.findall(r'world$', text)
print(match)  # ['world']
```

```python
text = "Hello aadsfasdfasfasdfa world"

pattern = r'^Hello .* world$'
rst = re.findall(pattern, text)

print(rst)
```

[]

의미: 문자 클래스 (대괄호 안에 있는 문자 중 하나)

```python
text = "apple banana cherry"

match = re.findall(r'[abc]', text)
print(match)  # ['a', 'b', 'a', 'a', 'a', 'c']
```

()

의미: 그룹화 (서브 패턴을 만들기 위해 사용)

```python
text = "abc abc abc"
match = re.findall(r'(abc)+', text)
print(match)  # ['abc', 'abc', 'abc']
```

?

의미: 0 또는 1회 반복

```python
text = "color colour"
match = re.findall(r'colou?r', text)
print(match)  # ['color', 'colour']
```

+

의미: 1회 이상 반복


```python
text = "123 4567 89"
match = re.findall(r'\d+', text)
print(match)  # ['123', '4567', '89']
```

*

의미: 0회 이상 반복

```python
text = "aba aaa aaaa"
match = re.findall(r'a*', text)
print(match)  # ['a', '', '', 'aaa', '', 'aaaa', '']
```

```python
import re

text = "사과123바나나456체리"
split_result = re.split(r"\d+", text)
print(split_result)
```

{n}

의미: 정확히 n회 반복

```python
text = "12345"
match = re.findall(r'\d{3}', text)
print(match)  # ['123']
```

{n, m}

의미: n회 이상 m회 이하 반복

```python
text = "12 123 1234 12345 1234567"
match = re.findall(r'\d{3,5}', text)
print(match)  # ['123', '1234', '12345']
```
